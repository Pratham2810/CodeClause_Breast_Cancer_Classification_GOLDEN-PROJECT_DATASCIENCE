```python
# the diagnosis column had the non number datatype as here we are excluding in this correlation or else we could have converted it
numeric_columns = df.select_dtypes(include='number').columns

# Calculate the correlation matrix
correlation_matrix = df[numeric_columns].corr()
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_colwidth', 10)   # Set the desired column width (e.g., 10)


# Print the correlation matrix
print(correlation_matrix)
```

```python
p=df[["radius_mean","texture_mean","smoothness_mean","compactness_mean","concavity_mean"]]
q=df["diagnosis"]
```


INORDER TO SPLIT THE TRAIN AND TEST SET

```python
from sklearn.model_selection import train_test_split
p_train,p_test,q_train,q_test=train_test_split(p,q,test_size=0.33)
```

LOGISTIC REGRESSION TO PREDICT

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(p_train,q_train)
predictions=logmodel.predict(p_test)
df1=p_test
df1
```

```python
len(p_test)
```

```python
predictions
```

KNN CLASSIFIER

```python
knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(p_train, q_train)
q_pred_test = knn.predict(p_test)
q_pred_train = knn.predict(p_train)
cm = confusion_matrix(q_test, q_pred_test)
acc_test = accuracy_score(q_test, q_pred_test)
acc_train = accuracy_score(q_train, q_pred_train)
score = knn.score(p_test, q_test)
print("Test Score: {}, Train Score: {}".format(acc_test, acc_train))
print("CM: ",cm)
```

```python
from sklearn.model_selection import GridSearchCV
knn_params = {'n_neighbors': np.arange(1,30,1)}
knn = KNeighborsClassifier()
knn_cv_model = GridSearchCV(knn, knn_params, cv=10)
knn_cv_model.fit(p_train,q_train)
```

```python
print("Best training score: {} with parameters: {}".format(knn_cv_model.best_score_, knn_cv_model.best_params_))
```

```python
from sklearn.neighbors import KNeighborsClassifier

import sklearn.metrics as metrics
model = KNeighborsClassifier(n_neighbors=6)
# Train the model using the training sets
model.fit(p_train,q_train)
y_pred=model.predict(p_test)
```

```python
print("Full Report",metrics.classification_report(q_test,y_pred))
```

SVM CLASSIFIER

```python
from sklearn import svm
import sklearn.metrics as metrics
clf = svm.SVC(kernel='linear') 
clf = clf.fit(p_train,q_train)
y_pred = clf.predict(p_test)
print("Full Report",metrics.classification_report(q_test,y_pred))
```
